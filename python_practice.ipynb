```python
def tri_recursion(x):
  if(x >0):
    result = x + tri_recursion(x-1)
  else:
    result = 0
  return result
print('recursion results')
tri_recursion(6)
```

```python
import pandas as pd
rewards_df = pd.read_csv('RewardsData.csv')
rewards_df.head()
```

```python
# creates a dataframe which contains either 'true' or 'false' to
# indicate whether there is missing data or not
rewards_df.isna()
```

```python
# counts total # of missing values in each column
missing_value_counts = rewards_df.isna().sum()
missing_value_counts
```

```python
# calculate % of missing values in each column
missing_value_perc = missing_value_counts/len(rewards_df)*100
missing_value_perc
```

```python
rewards_df['Last Seen']
```

```python
rewards_df['Last Seen'] = pd.to_datetime(rewards_df['Last Seen'])
rewards_df['Last Seen']
```

```python
rewards_df['Last Seen'].max()
```

```python
rewards_df['Last Seen'].min()
```

```python
rewards_df['Last Seen'].fillna(value = rewards_df['Last Seen'].min(), inplace = True)
```

```python
missing_value_counts = rewards_df.isna().sum()
print(missing_value_counts)
```

```python
# drop records with missing values, assign result to new df
rewards_df_sample = rewards_df.dropna()
rewards_df_sample
```

```python
# drop the 'tags' column from the sample df
rewards_df_sample = rewards_df_sample.drop(columns = 'Tags')
rewards_df_sample
```

```python
# unique values for the cities in the df
city_unique_values = rewards_df_sample['City'].unique()
city_unique_values

# sort these values
city_unique_values.sort()
```

```python
city_unique_values
```

```python
# there are several records of 'winston salem' with varying spelling
# these will need to be replaced

rewards_df_sample['City'].replace(to_replace=['Winston Salem', 'Winston Salem ',
       'Winston salem', 'Winston salem ', 'Winston-Salem',
       'Winston-Salem ', 'Winston-Salem, NC', 'Winston-salem'], value = 'Winston-Salem', inplace = True)
```

```python
rewards_df_sample['City'] = rewards_df_sample['City'].str.strip()
rewards_df_sample['City'] = rewards_df_sample['City'].str.title()
city_unique_values = rewards_df_sample['City'].unique()
```

```python
city_unique_values
```

```python
us_state_to_abbrev = {
   "Alabama": "AL",
   "Alaska": "AK",
   "Arizona": "AZ",
   "California": "CA",
   "Colorado": "CO",
   "Connecticut": "CT",
   "Delaware": "DE",
   "District of Columbia": "DC",
   "Florida": "FL",
   "Georgia": "GA",
   "Hawaii": "HI",
   "Idaho": "ID",
   "Illinois": "IL",
   "Indiana": "IN",
   "Iowa": "IA",
   "Kansas": "KS",
   "Kentucky": "KY",
   "Louisiana": "LA",
   "Maine": "ME",
   "Maryland": "MD",
   "Massachusetts": "MA",
   "Michigan": "MI",
   "Minnesota": "MN",
   "Mississippi": "MS",
   "Missouri": "MO",
   "Montana": "MT",
   "Nebraska": "NE",
   "Nevada": "NV",
   "New Hampshire": "NH",
   "New Jersey": "NJ",
   "New Mexico": "NM",
   "New York": "NY",
   "North Carolina": "NC",
   "North Dakota": "ND",
   "Ohio": "OH",
   "Oklahoma": "OK",
   "Oregon": "OR",
   "Pennsylvania": "PA",
   "Rhode Island": "RI",
   "South Carolina": "SC",
   "South Dakota": "SD",
   "Tennessee": "TN",
   "Texas": "TX",
   "Utah": "UT",
   "Vermont": "VT",
   "Virginia": "VA",
   "Washington": "WA",
   "West Virginia": "WV",
   "Wisconsin": "WI",
   "Wyoming": "WY"
}
```

```python
# replace each state name with abbreviated version in dictionary

for state_name in us_state_to_abbrev:
  rewards_df_sample['State'].replace(to_replace=state_name, value = us_state_to_abbrev[state_name], inplace = True)
```

```python
state_unique_values = rewards_df_sample["State"].unique()
state_unique_values.sort()
state_unique_values
```

```python
# convert us state dictionary to df for merging
states_df = pd.DataFrame(list(us_state_to_abbrev.items()),columns = ['State Name', 'State'])
states_df.head()
```

```python
# merge original df with the states df
rewards_df_sample = pd.merge(rewards_df_sample, states_df, on = 'State', how = 'left')
rewards_df_sample
```

```python
# rename state column to state abbrev
rewards_df_sample.rename(columns = {'State': 'State Abbreviation'}, inplace = True)
rewards_df_sample.head()
```

```python
# move state name column so that it shows up after state abbrev column

state_name_col = rewards_df_sample.pop('State Name')

rewards_df_sample.insert(3, 'State Name', state_name_col)
rewards_df_sample.head()
```

```python
zip_unique_values = rewards_df_sample['Zip'].unique()
zip_unique_values.sort()
```

```python
# string slicing
rewards_df_sample['Zip'] = rewards_df_sample['Zip'].str[:5]
```

```python
# boolean indexing to find zip codes that have 5 digits
zip_code_boolean_index = rewards_df_sample['Zip'].str.len() == 5
zip_code_boolean_index
```

```python
# drop records where the zips are < 5 digits
rewards_df_sample = rewards_df_sample[zip_code_boolean_index]
```
